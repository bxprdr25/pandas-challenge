```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#Calculate the total number of schools
total_schools = school_data["school_name"].count()

#Calculate the total number of students
total_students = student_data["student_name"].count()

#Calculate the total budget
total_budget = school_data["budget"].sum()

#Calculate the average math score
avg_math = student_data["math_score"].sum() / total_students

#Calculate the average reading score
avg_reading = student_data["reading_score"].sum() / total_students

#Calculate the percentage of students with a passing math score (70 or greater)
num_passing_math = ((student_data["math_score"]>=70).sum()/student_data["math_score"].count())*100

#Calculate the percentage of students with a passing reading score (70 or greater)
num_passing_reading = ((student_data["reading_score"]>=70).sum()/student_data["reading_score"].count())*100

#Calculate the percentage of students who passed math and reading (% Overall Passing)
overall_passing = (student_data[(student_data['math_score'] >= 70) & (student_data['reading_score'] >= 70)]['student_name'].count()/total_students)*100

#Create a dataframe to hold the above results
district_student_summary = pd.DataFrame({"Total Schools": total_schools,
                              "Total Students": total_students,
                              "Total Budget": total_budget,
                              "Average Math Score": avg_math,
                              "Average Reading Score": avg_reading,
                              "% Passing Math": num_passing_math,
                              "% Passing Reading": num_passing_reading,
                              "% Overall Passing": overall_passing}, index=[0])

#Format the dataframe for cleaner output
district_student_summary.style.format({"Total Students":"{0:,}", "Total Schools":"{0:,}", "Total Budget":"${:,.2f}", "% Passing Math":"{:,.2f}%",
                                      "% Passing Reading":"{:,.2f}%", "% Overall Passing":"{:,.2f}%"})
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
school_summary_df = school_data_complete.groupby(["school_name"])

#Get the school type
school_types_df = school_data.set_index("school_name")["type"]

#Calculate the total students by school
school_total_students = school_summary_df["student_name"].count()

#Calculate the total budget by school
school_budget = school_data.groupby(["school_name"]).budget.sum()

#Calculate the per student budget for each school
school_perStudent_budget = school_budget / school_total_students

#Calculate the average math score for each school
school_avg_math = school_summary_df["math_score"].mean()

#Calculate the average reading score for each school
school_avg_reading = school_summary_df["reading_score"].mean() 

#Calculat the % of students who passed math >= 70 for each school
school_percPass_math = (student_data[student_data["math_score"]>=70].groupby('school_name')['Student ID'].count()/school_total_students)*100

#Calculat the % of students who passed reading >= 70 for each school
school_percPass_reading = (student_data[student_data["reading_score"]>=70].groupby('school_name')['Student ID'].count()/school_total_students)*100

#Calculate the percentage of students who passed math and reading (% Overall Passing)
school_overall_passing = (student_data[(student_data['math_score'] >= 70) 
                     & (student_data['reading_score'] >= 70)].groupby('school_name')['Student ID'].count()/school_total_students)*100

#Create the summary dataframe for the schools
school_data_summary = pd.DataFrame({"School Type": school_types_df,
                              "Total Students": school_total_students,
                              "Total School Budget": school_budget,
                              "Per Student Budget": school_perStudent_budget,
                              "Average Math Score": school_avg_math,
                              "Average Reading Score": school_avg_reading,
                              "% Passing Math": school_percPass_math,
                              "% Passing Reading": school_percPass_reading,
                              "% Overall Passing": school_overall_passing})

#Apply styling to the summary dataframe for the schools
school_data_summary.style.format({"Total School Budget":"${:,.2f}","Per Student Budget":"${:,.2f}", "% Passing Math":"{:,.2f}%",
                                      "% Passing Reading":"{:,.2f}%", "% Overall Passing":"{:,.2f}%"})
```
